```python
import numpy as np;
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x, y):
    m_curr = b_curr = 0;
    iterations = 300
    
    learning_rate = 0.0001;
    cost_list=[]
    epoch_list=[]
    
    for i in range(iterations):
        
        y_predicted = m_curr * x + b_curr
        
        m_derivative = -sum(x * (y -  y_predicted))
        b_derivative = -sum(y -  y_predicted)
        
        m_curr = m_curr - learning_rate * m_derivative
        b_curr = b_curr - learning_rate * b_derivative
        
        cost = (1/2) * sum([val ** 2 for val in (y - y_predicted)])
        #print("m {}, iteration {}, cost {}".format(m_curr, b_curr, i, cost ))
        
        if(i%10==0):
            cost_list.append(cost)
            epoch_list.append(i)
    return cost_list, epoch_list


x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

cost_list, epoch_list = gradient_descent(x,y)
plt.xlabel("epoch")
plt.ylabel("cost");
plt.plot(epoch_list, cost_list)
```

```python
def logistic_regression_gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 300
    learning_rate = 0.0001
    cost_list = []
    epoch_list = []
    
    for i in range(iterations):
        y_predicted = 1 / (1 + np.exp(-(m_curr * x + b_curr)))
        
        m_derivative = -np.sum(x * (y - y_predicted))
        b_derivative = -np.sum(y - y_predicted)
        
        m_curr -= learning_rate * m_derivative
        b_curr -= learning_rate * b_derivative
        
        cost = -np.mean(y * np.log(y_predicted) + (1 - y) * np.log(1 - y_predicted))
        #print("m {}, b {}, iteration {}, cost {}".format(m_curr, b_curr, i, cost ))
        
        if i % 10 == 0:
            cost_list.append(cost)
            epoch_list.append(i)
    
    return cost_list, epoch_list

x = np.array([1, 2, 3, 4, 5])
y = np.array([0, 0, 1, 1, 1])

cost_list, epoch_list = logistic_regression_gradient_descent(x, y)
plt.xlabel("epoch")
plt.ylabel("cost")
plt.plot(epoch_list, cost_list)
plt.show()
```
