## Ques 12: Implement Linear Regression problem. For example, based on a dataset comprising of existing set of prices and area/size of the houses, predict the estimated price of a given house.

```python
#Required imports
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
# Reading csv file to dataframe
df = pd.read_csv('house.csv')
print(df.head())
# Scatter plot for the dataset
plt.xlabel('area')     # Setting x-axis label
plt.ylabel('price')    # Setting y-axis label
plt.scatter(df.area,df.price,color='green',marker='*')   # Scatter plot of area vs price
```

```python
reg = linear_model.LinearRegression()     #Creating a linear regression model object reg
reg.fit(df[['area']], df['price'])        #Fitting the model to the data, area as independent variable and price as dependent variable
m = reg.coef_                  # Getting the coefficient (slope) of the linear regression line
c = reg.intercept_             # Getting the intercept of the linear regression line
print('Coefficient, m = ', m)
print('Intercept, c = ', c)
p1 = reg.predict([[4444]])
print('(1) Price of a house with area = 4444 sqr ft: ', p1)
y = m*4444 + c
print('y = m*x + c =', y)
p2 = reg.predict([[7000]])
print('(2) Price of a house with area = 7000 sqr ft: ', p2)
y = m*7000 + c
print('y = m*x + c =', y)
x = np.linspace(2000,7500)
y = m*x+c
plt.plot(x, y, '-r', label='Best Fit Line')
plt.legend(loc='upper left')
plt.title('Graph of Best fit line')
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color="black",marker='*')
plt.grid()
plt.show()
```

```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def gradient_descent(X, y, learning_rate=0.01, num_iterations=1000):
    num_samples, num_features = X.shape
    weights = np.zeros(num_features)
    bias = 0

    for _ in range(num_iterations):
        z = np.dot(X, weights) + bias
        y_predicted = sigmoid(z)
        
        # Compute gradients
        dw = (1 / num_samples) * np.dot(X.T, (y_predicted - y))
        db = (1 / num_samples) * np.sum(y_predicted - y)
        
        # Update weights and bias
        weights -= learning_rate * dw
        bias -= learning_rate * db

    return weights, bias

# Example usage:
# Generate some random data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = np.random.randint(0, 2, size=(100,))

# Train the model using gradient descent
weights, bias = gradient_descent(X, y, learning_rate=0.01, num_iterations=1000)

# Make predictions
X_test = np.array([[1], [2]])
z_test = np.dot(X_test, weights) + bias
y_pred = sigmoid(z_test)
print("Predictions:", y_pred)
```
